```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np 
```

```python
deaths = pd.read_parquet('../../00_Data/Deaths_Data/export_data.parquet')
```

```python
deaths['DP_100k'] = deaths['Deaths'] / deaths['Population'] * 100_000
```

```python
annual_data = deaths.groupby(['State', 'Year'])[['DP_100k']].mean().reset_index()
```

```python
# Nebraska was qualitatively disqualified as a potential reference state
annual_data = annual_data[annual_data['State'] != 'NE']
```

## Washington

```python
colors = {
    'FL': 'b',
'IL':'g',
'KS':'tab:orange',
'OH':'c',
'OR':'m',
'TN':'y',
'TX':'tab:pink',
'WA':'k',
'WV':'0.5',
'WY':'tab:brown',
}

years = [2009,2010,2011,2012,2013]
wash_data = annual_data[annual_data['Year'].isin(years)]

# Set the style for the plot (optional)
plt.style.use('ggplot')

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plot each state's data as a separate line
for state in wash_data['State'].unique():
    if state == 'WA':
        state_data = wash_data[wash_data['State'] == state]
        ax.plot(state_data['Year'], state_data['DP_100k'], label=state, linewidth=3.0, color=colors[state], linestyle='dashed')
    else:
        state_data = wash_data[wash_data['State'] == state]
        ax.plot(state_data['Year'], state_data['DP_100k'], label=state, color=colors[state])


# Add titles and labels
ax.set_title('Washington: Drug-Related Deaths per 100k by State, Pre-2012')
ax.set_xlabel('Year')
ax.set_ylabel('Drug-Related Deaths per 100k')

ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))

ax.axvline(x=2012, color='r', linestyle='--', linewidth=1)

# Add a legend
ax.legend(title='State', loc='upper right', frameon=True, facecolor='0.9')

# Show the plot
plt.show()
```
